```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
test_ids = test["PassengerId"]
```

```python
def clean(data):
    data=data.drop(["Cabin",'Ticket','Name','PassengerId'],axis=1)
    cols=['Parch','SibSp','Age','Fare']
    for col in cols:
        data[col].fillna(data[col].median(),inplace=True)
    data.Embarked.fillna('U',inplace=True)
    return data

data=clean(data)
test=clean(test)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
data['Sex']=lc.fit_transform(data['Sex'])
print(lc.classes_)
data['Embarked']=lc.fit_transform(data['Embarked'])
print(lc.classes_)
```

```python
data.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
y=data['Survived']
x=data.drop('Survived',axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
xtrain,xval,ytrain,yval=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model=LogisticRegression(max_iter=1000,random_state=0)
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xval)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(yval,ypred))
```
