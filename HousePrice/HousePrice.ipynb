```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
test_ids = test["Id"]
```

```python
# check null values for every column
def check_null_values(data):
    null_count = data.isna().sum()  # Count nulls for each column
    for column, count in null_count.items():
        print(f" '{column}' : {count}")

check_null_values(data)
```

```python
data.shape
```

```python
#delete 2 columns cause morethan 90% columns are null
col = ['MiscFeature','Alley']
def delete_columns(data):
     for column in col:
        if column in data.columns:  # Check if column exists in the DataFrame
            data = data.drop(column, axis=1)
     return data

data=delete_columns(data)
```

```python
data.shape
```

```python
def fill_non_numeric_columns(data):
    for col in data.select_dtypes(exclude=['number']).columns:
        data[col]=data[col].fillna('missing value')
    return data

data=fill_non_numeric_columns(data)
```

```python
def fill_numeric_columns(data):
    for col in data.select_dtypes(include=['number']).columns:
        value=data[col].median()
        data[col]=data[col].fillna(value)
    return data

data=fill_numeric_columns(data)
```

```python
check_null_values(data)
```

```python
from sklearn.preprocessing import LabelEncoder
def label_encoding(data):
    lc=LabelEncoder()
    for col in data.select_dtypes(exclude=['number']).columns:
        data[col]=lc.fit_transform(data[col])
    return data
data = label_encoding(data)
```

```python
x = data.drop('SalePrice', axis=1)
y = data['SalePrice']
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
data.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data (fit on training data, transform both train and test)
scaler = StandardScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)

# Apply PCA (fit on training data, transform both train and test)
pca = PCA(n_components=0.95)  # Retain 95% of variance
xtrain_pca = pca.fit_transform(xtrain_scaled)
xtest_pca = pca.transform(xtest_scaled)

# Check the shapes
print("Original Train Shape:", xtrain.shape)
print("PCA Train Shape:", xtrain_pca.shape)
print("Original Test Shape:", xtest.shape)
print("PCA Test Shape:", xtest_pca.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, r2_score
regression = LinearRegression()
```

```python
regression.fit(xtrain_pca,ytrain)
```

```python
ypred = regression.predict(xtest_pca)
```

```python
print(r2_score(ytest,ypred))
print(root_mean_squared_error(ytest,ypred))
```
